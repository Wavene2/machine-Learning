```python
import pandas 
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pandas.read_csv('D:\machine learning\liner_reg\Salary_dataset.csv')

x=np.array(df["YearsExperience"],dtype=float)
y=np.array(df["Salary"],dtype=float)
plt.scatter(x,y)

m=0 
b=0
n=len(x)
epoch=1000
lr=0.0001 
for i in range(epoch):
    y_predict=m*x+b
    dm=(1/n)*np.sum((y_predict-y)*(x))
    db=(1/n)*np.sum(y_predict-y)
    m=m-lr*dm
    b=b-lr*db
    
y_predict=m*x+b
plt.plot(x,y_predict,color='red')
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Experience vs Salary")
plt.show()
print(m,b)
new_x=10
predicted_salary=m*new_x+b
print(f"Predicted salary for {new_x} years of experience is {predicted_salary}")


```
